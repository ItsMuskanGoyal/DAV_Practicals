```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q1
df = pd.DataFrame([[1, 1, 1], [2, 2, 2], [1, 2, 1], [2, 1, 1]],
                  index=['one', 'two', 'three', 'four'],
                  columns=pd.Index(['A', 'B', 'C'], name='MyPlot'))

# Create a figure and add two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot
sns.barplot(data=df, ax=axes[0])
axes[0].set_title('Bar Plot')
axes[0].set_xlabel('MyPlot')
axes[0].set_ylabel('Values')

# Horizontal bar plot
sns.barplot(data=df, ax=axes[1], orient='h')
axes[1].set_title('Horizontal Bar Plot')
axes[1].set_xlabel('Values')
axes[1].set_ylabel('MyPlot')

# Add legend
fig.legend(labels=df.columns, loc='upper right')

# Adjust layout
plt.tight_layout()
plt.suptitle('Bar and Horizontal Bar Plots', y=1.02)
plt.show()

# Q2
# Load iris dataset
iris = sns.load_dataset('iris')

# Scatter plot between Petal length and Sepal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='sepal_length', data=iris, label='Petal Length vs Sepal Length')

# Dashed line with red color
plt.axvline(x=3.5, color='red', linestyle='--', label='Threshold Line')

# Green marker with size 10 between Sepal length and Sepal width
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, color='green', marker='o', s=10, label='Sepal Length vs Sepal Width')

# Set labels and title
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.title('Scatter Plot: Petal Length vs Sepal Length')
plt.legend()
plt.show()
```
